```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

## Simple linear regression

```python
ndat = 10
delta = np.sqrt(ndat)

x = np.linspace(0, 9, ndat)
y = 2 * x + 3
y = y + np.random.uniform(-delta, delta, ndat)

df = pd.DataFrame({'x': x, 'y': y})
df.head()
```

```python
lm = LinearRegression()
lm.fit(df[['x']], df[['y']])
print('intercept: ', lm.intercept_ , 'coef: ', lm.coef_)
yhat = lm.predict(df[['x']])
```

```python
plt.scatter(df[['x']], df[['y']], label = 'experimental', color = 'red')
plt.plot(df[['x']], yhat, label = 'prediction')
plt.legend()
plt.show()
```

```python
df[['x', 'y']].plot()
plt.show()

df.plot(x = 'x', y = 'y')
plt.show()
```

```python
sns.regplot(x = 'x', y = 'y', data = df)
plt.show()
```

```python
sns.scatterplot(x = 'x', y = 'y', data = df)
plt.show()
```

## Polynomial regression

```python
ndat = 10
delta = np.sqrt(ndat)/50

x = np.linspace(0, 1, ndat)
y = 3 - 2*x +3*x**2
y = y + np.random.uniform(-delta, delta, ndat)

df = pd.DataFrame({'x': x, 'y': y})

plt.scatter(df[['x']], df[['y']], label = 'experimental', color = 'red')
plt.legend()
plt.show()
```

```python
poly_deg = 2
f = np.polyfit(x, y, poly_deg) 
p = np.poly1d(f)
df['yhat'] = p(df[['x']])
print(f)
print(p)
```

```python
plt.scatter(df[['x']], df[['y']], label = 'experimental', color = 'red')
plt.plot(df[['x']], df[['yhat']], label = 'prediction')
plt.legend()
plt.show()
```
